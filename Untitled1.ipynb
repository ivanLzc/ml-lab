```python
X = data['X4 number of convenience stores'].values
Y = data['Y house price of unit area'].values

#prosek x i y:
mean_x = np.mean(X)
mean_y = np.mean(Y)

# y = m*x + c
# m = m1/m2
n=len(X)
m1 = 0
m2 = 0
for i in range(n):
  m1 += (X[i] - mean_x) * (Y[i] - mean_y)
  m2 += (X[i] - mean_x) ** 2
m = m1 / m2
c = mean_y - (m * mean_x)

print (f'm = {m} \nc = {c}')

#crtanje grafika
max_x = np.max(X) + 100
min_x = np.min(Y) - 100

x = np.linspace (min_x, max_x, 100)
y = c + m * x

plt.plot(x, y, color='#3eb3ab', label='Regression Line')
plt.scatter(X, Y, c='#3a38b0', label='data points')

plt.xlabel('number of convenience stores')
plt.ylabel('house price of unit area')
plt.legend()
plt.show()

# R^2 se racuna da se vidi koliko je model dobar
# R^2 = r = r1/r2
r1 = 0
r2 = 0
for i in range(int(n)):
  y_pred = c + m * X[i]
  r2 += (Y[i] - mean_y) ** 2
  r1 += (Y[i] - y_pred) ** 2
r = 1 - (r1/r2)

print("R^2 = ", r)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = X.reshape(n, 1)
reg = LinearRegression()
reg = reg.fit(X,Y)

Y_pred = reg.predict(X)
r_square = reg.score(X, Y)

print ("R1^2 = ", r_square)
```
